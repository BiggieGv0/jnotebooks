```python
#Unzip, loading
import zipfile
import gzip
import pandas as pa

def load_df(filename):
    z = zipfile.ZipFile(filename, 'r')
    zf = z.namelist()
    file_object = z.open(zf[0], 'r')
    df = pa.read_csv(file_object, parse_dates=[2])
    return df

df0 = load_df('/home/mad/COT/deahistfo_1995_2012.zip')
df1 = load_df(path)
```

```python
c_en = 'Market and Exchange Names'
c_dt = 'As of Date in Form YYYY-MM-DD'
```

```python
df_b = df0.append(df1)
#df = df.sort(c_dt)  # does not work well
```

```python
#sorting
## ... "manual" sorting.
df = pa.DataFrame(
  sorted([v for i, v in df_b.iterrows()],
    key=lambda v: (v[c_dt], v[c_en].strip())),
  columns=df_b.columns)
#HTML(dfs.head(20).to_html())
```

```python
all_names = dict(enumerate(sorted(set(df[c_en]))))
#from pprint import pformat
with open('/home/mad/COT/exnames.lst', 'w') as f:
    f.write('\n'.join('%s: %s' % (k, v) for k, v in all_names.items()))
```

```python
## Filter out by single instrument
df_f = df[df[c_en] == col]
#HTML(df_f.head(100).to_html())
```

```python
## Group several instruments into one, Pandas-way.
import re
def igmap(iname):
    ## Grouping rules:
    ## Everything before " - "
    res = iname.split(' - ', 1)[0]
    ## ... without numbers
    res = re.sub(r'[0-9]', r'', res)
    ## ... without double spaces
    res = re.sub(r'  +', r' ', res)
    ## Without spaces in the beginning / end
    res = res.strip()
    #print "D: %r, %r" % (iname, res)
    return res
dfg = df.set_index(c_en).groupby(by=igmap)
# group = 'WHEAT' 
res = {}
for group in sorted(dfg.groups):
    res[group] = dfg.get_group(group).set_index(c_dt).groupby(level=0).sum()  #.head().to_html())
    res[group]['group'] = group
#dfg.set_index(c_dt)
dfg_a = pa.concat([res[k] for k in sorted(dfg.groups)])
```

```python
groups_enumerated = dict(enumerate(sorted(dfg.groups)))
## Print all the resulting instrument groups
with open('/home/mad/COT/grpnames.lst', 'w') as f:
    f.write('\n'.join('%s: %s' % (k, v) for k, v in groups_enumerated.items()))
## Print instrument groups and the underlying instruments
#dfg.groups
```

```python
dfg_a = pa.DataFrame(dfg_a, columns=(['group'] + [v for v in dfg_a.columns if v != 'group']))
```
