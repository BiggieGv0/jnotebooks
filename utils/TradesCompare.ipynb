```python
#!/usr/bin/env python
# coding: utf8


## PBPSWMeta
_META = dict(
    strategy_name='LDN',
    strategy_fn='london',
    security='RTS',
    tf=300,
    base_period=2,
)


import pybacktest
import pandas
import datetime


def london(ohlc):
    tz1 = 'Europe/Moscow'
    bars = ohlc.tz_localize(tz1)
    utcbars = bars.tz_convert('Europe/London')

    eve_t = '16:30'
    mor_t = '7:55'
    ex_t = '9:10'
    add_morning_time = lambda d: datetime.datetime.combine(d, datetime.time(7, 55, 0))

    ph = datetime.date(1900, 1, 1)

    mor = utcbars.C.at_time(mor_t).index
    mor = pandas.Series([i.date() for i in mor], index=mor)

    eve = utcbars.C.at_time('16:25')
    eve = pandas.Series(ph, index=eve.index)

    gr = eve.combine_first(mor)
    gr = gr.reindex(utcbars.index).bfill().fillna(value=ph)
    #return locals()

    H = utcbars.H.groupby(gr).max().iloc[1:]
    #.shift()#.iloc[1:].shift()
    #H.index = H.index.to_datetime()
    #H = H.reindex(utcbars.index, method='ffill').at_time(mor_t)

    L = utcbars.L.groupby(gr).min().iloc[1:]
    #.shift()#.iloc[1:].shift()
    #L.index = L.index.to_datetime()
    #L = L.reindex(utcbars.index, method='ffill').at_time(mor_t)

    C = utcbars.C.at_time(mor_t)
    C.index = [i.date() for i in C.index]

    #print L.tail(2)
    #print H.tail(2)
    #print C.tail(2)

    #C = C.reindex(L.index)

    #return locals()

    def idxfix(df):
        df = df.tz_convert(tz1)
        df.index = [i.replace(tzinfo=None) for i in df.index.to_pydatetime()]
        df = df.reindex(ohlc.index)
        df = df.fillna(value=False)
        return df

    buy = (C < ((H + 2*L) / 3))
    short = (C > ((2*H + L) / 3))

    utcix = pandas.DatetimeIndex(buy.index.map(add_morning_time))
    utcix = utcix.tz_localize('Europe/London')

    buy.index = utcix
    short.index = utcix

    #return locals()


    dn = pybacktest.DateNum(utcbars)
    exits = (utcbars.C.at_time(ex_t).astype(bool))
    exits = exits.combine_first(dn != dn.shift(1))
    exits = idxfix(exits)
    sell = cover = exits
    buy = idxfix(buy)
    short = idxfix(short)
    return locals()
    #return dict(ohlc=ohlc, Buy=buy, Sell=sell, Short=short, Cover=cover)
```
