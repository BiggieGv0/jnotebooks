```python
## Supposed StopLoss optimization
data_in = data_f; x_col = 'MAEp'; y_col = 'Profit'
res = float("inf")
x_all, y_all = data_in[x_col], data_in[y_col]
res = []
for x, y in sorted(zip(x_all, y_all)):
    worse = x_all < x
    y_v = y_all.copy()
    y_v[worse] = x
    r = y_v.sum()
    res.append((x, r))

pandas.DataFrame(res, columns=('SL', 'Profit')).set_index('SL').plot()
```
