```python
def intersect_index(*dataframes):
    common_index = set.intersection(*(set(df.index) for df in dataframes))
    common_index = sorted(common_index)
    ## XX: alternatively: iterate over the index of the first dataframe filtering out the non-common
    #common_index = [v for v in dataframes[0].index if v in common_index]
    return tuple(df.reindex(common_index) for df in dataframes)
```

```python
#defining bollinger bands indicator, where ts = timeseries to use, 
#win = window of rolling mean and std and n = std multyplier
def bbands(ts, win, n):
    std = pa.rolling_std(ts, win)
    bband = pa.rolling_mean(ts, win)
    bbandTop = bband + std*n
    bbandBot = bband - std*n
    return bband, bbandTop, bbandBot

#getting ADX indicator timeseries
def adx(ser):
    adx = ADX_iter(ser, 5).ADX
```

```python
from __future__ import division
def C_Raptor(ohlc, spx):

    ## Alises AMI -> Pandas
    _min_periods = 1
    MA = lambda s, l: pa.rolling_mean(s, l, min_periods=_min_periods)
    LLV = lambda s, l: pa.rolling_min(s, l, min_periods=_min_periods)
    StDev = lambda s, l: pa.rolling_std(s, l, min_periods=_min_periods)
    Ref = lambda series, num: series.shift(-num)
    log = np.log
    Min = lambda *serieses: pa.concat(serieses, axis=1).min(axis=1)
    from pyquant.transforms import RSI as q_RSI, RSI_iter as q_RSI_iter
    RSI = lambda n: q_RSI_iter(ohlc.C, n)
    O, H, L, C, V = ohlc.O, ohlc.H, ohlc.L, ohlc.C, ohlc.V

    ##### НАСТРОЙКИ СКАНЕРА ЦЕНЫ и ЛИКВИДНОСТИ
    ## 1. определение ликвидности бумаги
    ## определение среднего значения V * C у индекса за 65 дней. ВНИМАНИЕ!!! Некотораые поставщики
    ## данных не даеют значение "объем" у индекса
    #SetForeign("$SPX")  ## обращение к стороннему индексу
    O, H, L, C, V = spx.O, spx.H, spx.L, spx.C, spx.V
    VolSP = MA(V * C, 65)
    #RestorePriceArrays()
    O, H, L, C, V = ohlc.O, ohlc.H, ohlc.L, ohlc.C, ohlc.V

    xvolFilter = 3.0 / 1000000
    reqs = MA(C * V, 20) > (VolSP * xvolFilter)

    ## 2.Расчет исторической волы и её оценка
    HV = 100 * StDev(log(C / Ref(C, -1)), 100) * sqrt(252)
    histVola = HV > 20

    ## 3.Проверка на наличие макротренда
    bigTrend = C > MA(C, 200)

    ## 4.Проверка на наличие отката
    otkatPrice = C < MA(C, 20)

    ## 5.Проверка на резкость отката
    hardRSI = RSI(2) < 5

    ## 6.опять же оцениваем резкое падение вчера
    hardOtkat = ((C < Ref(C, -5) * 0.85) | (C < Ref(C, -2) * 0.925))

    ## 7.факультатив - цена должна быть больше 5 баксов. Сделано чтобы отсечь неликвид. НО! чисто визуально отобранные выше стоки < 5 выглядят вполне себе ликвидными
    cutPrice = C > 2

    BuyHR = reqs & histVola & bigTrend & otkatPrice & hardRSI & hardOtkat & cutPrice

    #####  ОПРЕДЕЛЕНИЕ УРОВНЯ ЛИМИТНИКА
    ### если сканер вчера посканировал и нашел акции то перед открытием ставим лимитники.
    ### ВНИМАНИЕ!!! АМИ под словом "вчера" понимает день, предшевствовавший выставлению лимитника или просто предыдущий бар независимо от того
    ## какой ТФ используется

    Lmt = 8  ## уровень лимита в процентах
    PriceTrigger = Ref(C, -1) * (1 - Lmt / 100)  ## расчет уровня в цене относительно вчерашнего закрытия
    Trigger = L < PriceTrigger   ## если минимум бара ниже лимитника - входим
    bp = Min(O, PriceTrigger)  ## определение цены входа - случай ДЛЯ АМИ описывает ситуацию когда опен оказался ниже лимитника, то входим по опену - чисто для имитации реала
    Buy = Ref(BuyHR, -1) & (Trigger)  ## собираем все вместе

    BuyPrice = bp

    ################################################################################
    ##  Sell Rules
    RSI_2 = RSI(2)
    Sell = Ref(RSI_2, -1) > 70  ## если с момента входа прошел минимум один день и RSI(2) > 70 выходим
    SellPrice = O

    #########################################
    Short = Cover = pa.Series(False, index=O.index)

    ## СТОПОВ НЕ ПРИДУМАЛ - ДУМАЮ ДИВЕРСИФИКАЦИЯ СПАСАЕТ МИР

    Buy_b = Buy
    Buy = Ref(Buy, 1)
    Sell_b = Sell
    Sell = Ref(Sell, 1)
    BuyPrice_b = BuyPrice
    #BuyPrice = Ref(BuyPrice, 1)
    # BuyPrice = pa.Series(1.234, index=O.index)
    buyprice = BuyPrice
    ## XXXXX:
    buyprice = BuyPrice = sellprice = SellPrice = C

    return locals()

loc = C_Raptor(ohlc, spx)
```

```python
def run_bt(symbol, bench, strat=C_Raptor, start=dt.datetime(1999, 1, 1)):
    #sets starting date of backtest

    #getting data of Stock OHLC
    OHLC = marketdata.ext.load_metotron(symbol)
    OHLC = OHLC.ix[start:]

    #getting data of SP500 index
    SPX = marketdata.ext.load_metotron(bench)
    SPX = SPX.ix[start:]
    
    OHLC, SPX = intersect_index(OHLC, SPX)

    #getting buy, sell, short, cover signals vectors
    res = strat(OHLC, SPX)

    #getting backtest object
    bct = pbt.Backtest(dict(res, ohlc=OHLC), name=repr(strat))

    #getting perfomance report and profit factor
    perf = bct.report
    #pf = perf['performance']['PF']
    
    return res, bct

sym = _sym
sig, bct = run_bt(sym, '$SPX')
bct.plot_equity()
bct.summary()
bct.trades.to_csv('/home/data/tmp/tmpf2.csv')
bct.trades
```
