```python
#defining bollinger bands indicator, where ts = timeseries to use, 
#win = window of rolling mean and std and n = std multyplier
def bbands_xxx(ts, win, n):
    """ bollinger bands indicator.
    :param ts: timeseries to use
    :param win: window of rolling mean and stddev
    :param n: stddev multiplier
    """
    std = pa.rolling_std(ts, win)
    bband = pa.rolling_mean(ts, win)
    bbandTop = bband + std * n
    bbandBot = bband - std * n
    return bband, bbandTop, bbandBot

def bbands(series, win, n):
    bb = pyq_bbands(win, n)
    for idx, val in series.iteritems():
        bb.add(val, timestamp=idx)
    res = bb.series
    return res['center'], res['upper'], res['lower']

#getting ADX indicator timeseries
def adx(ser):
    adx = ADX_iter(ser, 5).ADX
    return adx
```

```python

def baseCB(ohlc, spx):

    ## Helpers / aliases / settigns
    Min = lambda *serieses: pa.concat(serieses, axis=1).min(axis=1)
    _min_periods = 1
    MA = lambda s, l: pa.rolling_mean(s, l, min_periods=_min_periods)
    LLV = lambda s, l: pa.rolling_min(s, l, min_periods=_min_periods)

    O, H, L, C, V = ohlc.O, ohlc.H, ohlc.L, ohlc.C, ohlc.V

    # SCANNER #
    #liquidity filter
    volSP = MA(spx.V * spx.C, 65)
    xvolFilter = 3.0 / 1000000  # minimum threshold
    liqIndex = MA(C * V, 20) > (volSP * xvolFilter)

    #getting bbands indicator series
    bband, bbandTop, bbandBot = bbands(C, 5, 1)
    
    #defining C in range of BbandTop - BbandBot
    b = (C - bbandBot) / (bbandTop - bbandBot)
    
    #price filter
    minPrice = C > 5 #minimum threshold
    maxPrice = C < 100 #maximum threshold
    
    #'price & liq' SCANNER
    priceLiqScan = liqIndex & minPrice & maxPrice
    
    #trend filter
    trendFilt = C > MA(C, 200)
    
    #'force of trend' SCANNER
    forceFilt = adx(ohlc) > 40.0
    
    #volume filter
    volFilt = V > LLV(V, 5)

    #defining filter "deep pullback of b"
    pullback = (pbt.BarsSince(b > 0.5) > 3.0) & (b <= 0.1)
    
    #price action scanner #SetupScan
    priceScan = trendFilt & forceFilt & volFilt & pullback
    
    #final SCANNER #BuyHR
    setupScan = priceLiqScan & priceScan
    
    # TRIGGER #
    lmt = 3.0
    PriceTrigger = C.shift(1) * (1 - lmt / 100)
    trigger = (L < PriceTrigger)
    
    # SIGNALS #
    buy = setupScan & trigger
    buy_price = Min(O, PriceTrigger)
    sell1 = C > MA(C, 3)
    sell2 = h.nbarexit(buy, 5)
    sell = sell1 | sell2
    short = pa.Series(0, index=ohlc.index)
    cover = pa.Series(0, index=ohlc.index)
    #res = dict(Buy=buy, Sell=sell, Short=short, Cover=cover)
    #return res
    return locals()
    
```

```python
def CB(symbol, bench):
    #sets starting date of backtest
    start = dt.datetime(2000, 1, 1)
    
    #getting data of Stock OHLC
    OHLC = marketdata.ext.load_metotron(symbol)
    OHLC = OHLC.ix[start:]
        
    #getting data of SP500 index
    SPX = marketdata.ext.load_metotron(bench)
    SPX = SPX.ix[start:]
    
    #getting buy, sell, short, cover signals vectors
    res = baseCB(OHLC, SPX)
    
    #getting backtest object
    bct = pbt.Backtest(dict(res, ohlc=OHLC), name = 'Base CB str')
    
    #getting perfomance report and profit factor
    perf = bct.report
    #pf = perf['performance']['PF']
    
    return res, bct
```
