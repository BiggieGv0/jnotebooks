```python
kl = 1.2
wind = 3

param_wd = 'Thu'
d = dohlc_a
sd = pandas.rolling_apply(dohlc.C, wind, numpy.std).reindex(dohlc.index)
up = sd * kl
mon = d[d['wd'] == 'Mon']
mon = mon.reindex(dohlc_a.index).ffill()
fri = d[d['wd'] == 'Fri'].reindex(dohlc_a.index).ffill()
p_wd = d[d['wd'] == param_wd].reindex(dohlc_a.index).ffill()
## TODO: skip the whole weeks where there is no mondex
## XXX: value is irrelevant on mondays.
diff = dohlc_a['td'] = p_wd.O - mon.O
diff = diff.reindex(dohlc.index)
diff_a = diff.abs()
istu = dohlc_a['wd'] == param_wd
buy = (diff < 0) & (diff_a > up) & istu & #((fri.shift(5).C - mon.shift(5).C) < 0)
short = (diff > 0) & (diff_a > up) & istu
cover = sell = dohlc_a['wd'] == 'Fri'
ret = dict(Buy=buy, Short=short, Sell=sell, Cover=cover, ohlc=dohlc)
```
