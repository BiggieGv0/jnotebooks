```python
import pandas as pd
from marketdata import ccomb
import common_modules
import matplotlib.pyplot as plt
#reload(common_modules)
```

```python
Symbol = 'RTS'
TF = 'bar_timeframe_900'
df = common_modules.pb_data_loading(Symbol, TF)
```

```python
df['Volat'] = df.H - df.L
#df['Diff'] = df.Volat.diff()
df['Mean'] = pd.rolling_mean(df.Volat, 3)
df['Std'] = pd.rolling_std(df.Volat, 15)
df['UpBand'] = df.Mean + df.Std
df['DwnBand'] = df.Mean - df.Std
df['VolBreak'] = df.Volat > df.UpBand 
```

```python
df2 = df.ix[df.VolBreak > 0][3000:]
plt.figure(figsize=(25,6))
plt.subplot(211)
plt.xlabel('-')
plt.ylabel('Points')
plt.title('Volatility breakdowns')
plt.grid(True)
plt.plot(df.Volat, 'g-')
plt.plot(df.Mean, 'r-')
plt.plot(df.UpBand, '-')
```
