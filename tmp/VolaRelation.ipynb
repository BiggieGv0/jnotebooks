```python
#BA strategy
ba['volat'] = (ba.H - ba.L)/ba.C

"""
Percent Range
"""

def decil(rang, val):
    total = bigger = 0
    for x in rang:
        total += 1
        if x <= val:
            bigger += 1
    decil = float(bigger)/total
    return decil

def fta(series):
    return decil(series[:-1], series[-1])

ba['Decile'] = pa.rolling_apply(ba.volat, 50, fta)
#ba['PreVol'] = pa.rolling_mean(ba.volat, 5)
#ba['PostVol'] = pa.rolling_mean(ba.volat.shift(-5), 5)

#ba['VolaRelation'] = np.log(ba['PostVol'] / ba['PreVol'])
ba['Buy'] = (ba.Decile >= 0.4) & (ba.Decile < 0.5)
#ba_cut = ba.ix['2012-01-01':]
#going to next days Open 
#expiration_dates = set(
#    simple_expiration_date(v, from_last_month_day=False, skip_days=3)
#    for v in ba_cut.index)
#ba['Expiries'] = ba.index.map(lambda v: v in expiration_dates)
#ba['Exit'] = ba.Expiries
#ba['Buy'] = ExRem(ba.Cond1, ba.Exit)
#ba.ix['2013-12-01':]


ba['PostVol'] = None
for idx, row in ba[ba['Buy']].iterrows():
    _, _, opt_b = get_options(symbol='SPY', date=idx.date(),
                              price=row['C'], opt_type='C', distance=0)
    postvol = ba.ix[idx:opt_b.name]['C'].mean()
    ba.ix[idx, 'PostVol'] = postvol

#def PostVolat(current index, target index):
#    for i in index range:
#        calc of mean
#    return mean

```

```python
#Filtering only buy signals
pd2 = ba.copy()
pd2['Profit'] = 0
pd2 = pd2.ix['2012-01-01':]
#pd2
```

```python
#Options strategy
_res = []
import functools
option_profit_my = functools.partial(option_profit, midstrike_rounding='by_type')
import pdb
import traceback
import sys


def optprofit2data(optprofit_ret, prefix=''):
    if not optprofit_ret:
        return dict()
    res = dict(
        strike=optprofit_ret['optinfo']['strike'],
        price=optprofit_ret['opt_price'],
        profit=optprofit_ret['profit'],
    )
    if prefix:
        res = {'%s%s' % (prefix, k): v for k, v in res.items()}
    return res


debug = 3

#relevant_index = [pd2.ix['2013-12-30':].index[0]]
#  Normally:
relevant_index = pd2.index

for idx in relevant_index:
    row = pd2.ix[idx]
    if not row['Buy']:
        continue
    try:
        next_day = pd2.shift(-1).ix[idx]
    except KeyError:
        print "No next day for %r" % (idx,)
        continue
    if debug >= 3:
        print "\nidx:", idx
        print "\nrow:", row.to_dict()
        print "\nnext_day:", next_day.to_dict()
        print "\n"
    price = Decimal(next_day.O)
    _p = dict(
        price=price,
    )
    _p1 = _p2 = _p3 = _p4 = None
    kwa = dict(
        symbol='SPY',
        date=row.name.date(),  #idx.date(),
        price=price,
        debug=debug,
        return_all=True,
        base_active=ba
    )
    try:
        _p1 = option_profit_my(distance=0, direction=-1, opt_type='C', **kwa)
        _p2 = option_profit_my(distance=4, direction=1, opt_type='C', **kwa)
        _p3 = option_profit_my(distance=0, direction=-1, opt_type='P', **kwa)
        _p4 = option_profit_my(distance=-4, direction=1, opt_type='P', **kwa)
    except NoData as exc:
        print "NoData at %r: %r" % (idx, exc)
        continue

    _p.update(
        close_date=_p1['close_date'],
        ba_close_price=_p1['ba_on_close'].C,
        ba_on_close_date=_p1['ba_on_close'].name,
    )
    _p.update(optprofit2data(_p1, prefix='o1_'))
    _p.update(optprofit2data(_p2, prefix='o2_'))
    _p.update(optprofit2data(_p3, prefix='o3_'))
    _p.update(optprofit2data(_p4, prefix='o4_'))

    if _p1 and _p2 and _p3 and _p4:
        _p.update(
            profit=_p1['profit'] + _p2['profit'] + _p3['profit'] + _p4['profit'],
        )

    if debug >= 3:
        print "\nres line:", _p
    _res.append((idx, _p))

print "I: results length: %r" % (len(_res),)
_profits = pa.DataFrame([v for k, v in _res], index=[k for k, v in _res])

pd2['close_date'] = _profits['close_date']
pd2['ba_on_close'] = _profits['ba_close_price']
#pd2['req_strike_price'] = _profits['price']
pd2['call_ATM_strike'] = _profits['o1_strike']
pd2['call_ATM_price'] = _profits['o1_price']
pd2['call_ATM_profit'] = _profits['o1_profit']
pd2['call_OTM_strike'] = _profits['o2_strike']
pd2['call_OTM_price'] = _profits['o2_price']
pd2['call_OTM_profit'] = _profits['o2_profit']
pd2['put_ATM_strike'] = _profits['o3_strike']
pd2['put_ATM_price'] = _profits['o3_price']
pd2['put_ATM_profit'] = _profits['o3_profit']
pd2['put_OTM_strike'] = _profits['o4_strike']
pd2['put_OTM_price'] = _profits['o4_price']
pd2['put_OTM_profit'] = _profits['o4_profit']
pd2['sum_profit'] = _profits['profit']
```

```python
#Finalizing ba strategy and save it to excel
pd2f = pd2[pd2['Buy']]
pd3 = pd2f.ix['2012-01-01':'2014-01-01']
pd3 = pd3.drop(['Profit', 'Decile', 'volat','Buy','H', 'L'], 1)
pd3['CumSumm'] = pd3.sum_profit.cumsum()
pd3.to_excel('/home/mad/Options/Results/SPY/Backtests/batterfly_on_4_decile.xlsx')
```
