```python
"""
Contract expiration date (experiments / helpers)
"""
import datetime
import calendar


def date_add_months(sourcedate, months=1):
    month = sourcedate.month - 1 + months
    year = sourcedate.year + month / 12
    month = month % 12 + 1
    day = min(sourcedate.day, calendar.monthrange(year, month)[1])
    # return datetime.date(year,month,day)
    return sourcedate.replace(year=year, month=month, day=day)


def date_month_last_day(date):
    # next_month = date_add_months(date, 1).replace(day=1)
    # return next_month - datetime.timedelta(1)
    return date.replace(day=calendar.monthrange(date.year, date.month)[1])


## TODO: skipping holidays;
## however, it is unclear whether some saturdays/sundays can be business days.
def simple_expiration_date(
        month_date, skip_days, holidays=True,
        from_last_month_day=True, delta=-1, debug=False):
    """ ... """

    if holidays is True:
        holidays = holidays_list  ## from global
    
    current = month_date
    if from_last_month_day:
        last_day = date_month_last_day(month_date)
        if debug:
            print "last month day:", last_day
        current = last_day

    business_days_skipped = 0
    while True:  # business_days_skipped < skip_days:
        ## TODO?: holidays correct list for all the relevant time
        if current.isoweekday() in (6, 7):  ## saturday / sunday
            if debug:
                print "weekend (%s): %s" % (current.strftime("%A"), current)
        elif holidays and current in holidays:  ## holiday
            if debug:
                print "holiday from list: %s" % (current,)
        else:
            if debug:
                print "business day:", current
            ## Found a weekday we can return
            if business_days_skipped >= skip_days:
                break
            business_days_skipped += 1
        current += datetime.timedelta(delta)

    return current


if _run_dbg:
    print "2013-01-04?:", simple_expiration_date(
        datetime.date(2013, 1, 6), skip_days=0,
        from_last_month_day=False, debug=True)  ## 
    print ""
    # print "2007-07-25?:", simple_expiration_date(datetime.date(2007, 7, 1), skip_days=3, debug=True)
    print "2010-05-25?:", simple_expiration_date(datetime.date(2010, 5, 1), skip_days=3, debug=True)
```

```python
"""
Option parsing helpers
"""


def find_files(in_dir, fname_re=None):
    """ Return all full file paths under the directory `in_dir` whose
    *filenames* match the `fname_re` regexp (if not None) """
    for dir_name, dir_list, file_list in os.walk(in_dir):
        if fname_re is not None:
            file_list = (v for v in file_list if re.match(fname_re, v))
        ## Annotate with full path:
        file_list = ((os.path.join(dir_name, file_name), file_name) for file_name in file_list)
        for fpath, fname in file_list:
            yield fpath, fname


## Base set of keys common to all option sources (abstraction format, basically)
option_common_keys = set((
    'file_path symbol exp_date opt_type strike load_func'
).split())


def filter_option_data(data, postfilter=True):
    if postfilter or postfilter is None:  ## need 'False' to disable
        if hasattr(postfilter, '__iter__'):
            keys = set(postfilter)
        else:
            keys = option_common_keys
        data = dict((k, v) for k, v in data.items() if k in keys)
    return data


def ohlcv_resample(data, rule='D', fields=('O', 'H', 'L', 'C', 'V')):
    return pandas.DataFrame(dict(zip(
        fields,
        [data[fields[0]].resample(rule, how=lambda x: x[0] if len(x) > 0 else None),
         data[fields[1]].resample(rule, how=lambda x: x.max()),
         data[fields[2]].resample(rule, how=lambda x: x.min()),
         data[fields[3]].resample(rule, how=lambda x: x[-1] if len(x) > 0 else None),
         data[fields[4]].resample(rule, how=lambda x: x.sum())]
        )),
      columns=fields  # for default columns order
     ).dropna()
```

```python
class OptFileParser(object):
    """ The option filename and data parser common (abstract) class """

    ## To be overridden:
    fln_re = re.compile(r'')
    load_func = None

    ## Temporary variables (for extensibility) (be careful with the lifetime)
    current_data = None
    kwa = None

    class ParseError(ValueError):
        """ A parser-specific exception for 'could not parse that' """

    def __init__(self, postfilter=True):
        self.postfilter = postfilter

    def try_parse(self, *ar, **kwa):
        try:
            return self.parse_or_raise(*ar, **kwa)
        except self.ParseError as exc:
            return exc.args[0]
        finally:
            self._cleanup()

    parse = try_parse

    __call__ = try_parse

    def _cleanup(self):
        self.current_data = None
        self.kwa = None

    def parse_or_raise(self, file_path, **kwa):
        self.kwa = kwa  ## Hey, extra info!
        filename = os.path.basename(file_path)
        pre_data = dict(
            filename=filename,
            file_path=file_path,
        )
        self.current_data = pre_data

        data = self.parse_filename(filename)
        data.update(pre_data)
        self.current_data = data

        ## Postprocess the match:
        data.update(self.parse_exp_date(data))
        data.update(self.parse_strike(data))
        data.update(self.get_load_func(data))
        data.update(self.parse_opt_type(data))
        data.update(self.get_extra_data(data))

        for key_ in option_common_keys:
            assert key_ in data, "%s should be in the resulting data" % (key,)
            assert callable(data['load_func']), "load_func should be a callable"

        data = self.filter_option_data(data)

        return data

    """
    Sample filtered data:
    {'exp_date': datetime.date(2012, 8, 27),
     'file_path': '/home/mad/Options/QNG/QNGQ12P1750_D.csv',
     'opt_type': 'P',
     'strike': '1750',
     'symbol': 'QNG'}
    """

    def filter_option_data(self, data):
        postfilter = self.kwa.get('postfilter', self.postfilter)
        return filter_option_data(data, postfilter=postfilter)
    
    def parse_filename(self, filename):
        match = self.fln_re.match(filename)
        if not match:
            return self._mk_error('unmatched', regex=self.fln_re.pattern)
        return match.groupdict()

    def _mk_error(self, error, data=None, **kwa):
        exc_data = dict(data or self.current_data)
        exc_data.update(kwa)
        exc_data.update(error=error)
        exc = self.ParseError(exc_data)
        self._cleanup()
        raise exc

    def parse_exp_date(self, data):
        raise NotImplementedError("Abstract class")

    def parse_strike(self, data):
        return dict(strike=ftype(data['strike']))

    def parse_opt_type(self, data):
        return {}  ## Assume it's in the data alraedy

    def get_load_func(self, data):
        return dict(load_func=self.load_func)

    def get_extra_data(self, data):
        return {}

    def nearest_expiration_date(self, date):
        raise NotImplementedError  ## NOTE: does not have to be overridden!

    
class DummyParser(OptFileParser):
    pass


def load_from_fileinfo_cached(fileinfo):
    data = fileinfo.get('data')
    if data is None:
        filepath = fileinfo['file_path']
        load_func = fileinfo['load_func']
        data = load_func(filepath, fileinfo=fileinfo)
        fileinfo['data'] = data
    return data
```

```python
"""
CME
"""


class OptParserCME(OptFileParser):
    fln_re = re.compile((
        r'^'  ## filename, from the start (could use '(?:^|/)' though)
        r'(?P<symbol>[A-Z0-9a-z@+]+?)'
        r'(?P<exp_month>[FGHJKMNQUVXZ])'
        r'(?P<exp_year>[0-9]{2})'
        r'(?P<opt_type>[PC])'
        r'(?P<strike>[0-9.]+)'
        r'(?P<extra>.*?)'  ## The remainder
        r'$'  ## Up to the end (explicit)
    ))

    ## filename's exp_month to the month number
    ## http://www.iqfeed.net/symbolguide/index.cfm?symbolguide=guide&displayaction=support&section=guide&web=iqfeed&guide=commod&web=IQFeed&symbolguide=guide&displayaction=support&section=guide&type=nymex&type2=nymex_gbx&type3=nymexmini
    exp_month_mapping = {
        'F': 1, 'G': 2, 'H': 3, 'J': 4, 'K': 5, 'M': 6,
        'N': 7, 'Q': 8, 'U': 9, 'V': 10, 'X': 11, 'Z': 12}

    """
    Example `(filename, data)`:
     ('QNGG12P3250_D.csv',
      {'exp_month': 'G',
       'exp_year': '12',
       'extra': '_D.csv',
       'opt_type': 'P',
       'strike': '3250',
       'symbol': 'QNG'})]
    """

    def parse_exp_date(self, data):
        ## regex should make sure failures don't happen here
        exp_month_num = self.exp_month_mapping[data['exp_month']]
        exp_month_num = int(exp_month_num)

        exp_year = data['exp_year']
        if len(exp_year) == 2:
            exp_year = '20%s' % (exp_year,)
        exp_year = int(exp_year)

        exp_month_date = datetime.date(exp_year, exp_month_num, 1)
        ## NOTE: using the previous month, apparently
        ## http://www.energysolutionsinc.com/ace-files/NYMEX_Calendar.pdf
        exp_month_date = date_add_months(exp_month_date, -1)
        exp_date = simple_expiration_date(exp_month_date, skip_days=3)

        return dict(
            exp_date=exp_date,
            ## Debug stuff
            exp_month_num=exp_month_num,
            exp_year_parsed=exp_year,
            exp_month_date=exp_month_date,
        )

    def parse_strike(self, data):
        strike = data['strike']
        strike = ftype(strike) / 1000  # to match the BA price
        return dict(strike=strike)

    _columns = 'timestamp O H L C V OI'.split()
    _target_columns='O H L C V OI'.split()

    @classmethod
    def load_func(
            cls,
            filepath,
            ## Possibly wrong
            columns=None,
            ## Desired ordering (for output)
            target_columns=None,
            fileinfo=None):
        df = pa.read_csv(
            filepath,
            index_col=0,  #'timestamp',
            parse_dates=True,
            date_parser=lambda v: datetime.datetime.strptime(v, '%d.%m.%Y'),
            header=None,
            names=columns or cls._columns,
            # skip_rows=0,
        )
        df = pa.DataFrame(df, columns=target_columns or cls._target_columns)
        # df = ohlcv_resample(df)
        return df

##    TODO:
#     def nearest_expiration_date(self, date):
#         pass


if _run_dbg:
    print OptParserCME().parse(base_dir + 'CME/QNG/QNGM09P4700_D.csv')
    print ""
    print OptParserCME.load_func(base_dir + 'CME/QNG/QNGM09P4700_D.csv').head(2)
```

```python
def load_fileinfos(data_dir, func, **kwa):
    ## NOTE: fname_re is from the beginning of the filename
    files = list(find_files(data_dir, r'(?i).*\.csv$'))

    fileinfos = [
        func(file_path=fpath, **kwa)
        for fpath, fname in files]
    return fileinfos


def load_metadata(*ar, **kwa):
    ## The whole option-manager
    mgr = kwa.pop('mgr', None)
    if mgr is not None:
        kwa.setdefault('func', mgr.parse)
    else:
        mgr = DummyParser()

    fileinfos = load_fileinfos(*ar, **kwa)
    debug = kwa.get('debug', False)

    db = kwa.get('db') or {}
    for fileinfo in fileinfos:
        if 'error' in fileinfo:
            if debug:
                print "Non-loaded file:", fileinfo
            continue
        symbol_db = db.setdefault(fileinfo['symbol'], {})
        symbol_db.setdefault('__manager', mgr)
        assert type(symbol_db['__manager']) is type(mgr)
        date_db = symbol_db.setdefault(fileinfo['exp_date'], {})
        dir_db = date_db.setdefault(fileinfo['opt_type'], {})
        dir_db[fileinfo['strike']] = fileinfo

    return db


def load_metadata_cached(*ar, **kwa):
    global data_db
    if data_db is None:
        data_db = load_metadata(*ar, **kwa)
    return data_db

if _run_dbg:
    import random
    print "QNG:"
    print pretty(random.sample(
        load_fileinfos(base_dir + 'CME/', func=OptParserCME().parse),
        5))
```

```python
"""
CBOE
"""


class OptParserCBOE(OptFileParser):

    fln_re = re.compile((
        r'^'  ## filename, from the start (could use '(?:^|/)' though)
        r'(?P<symbol>[A-Z0-9a-z@+]{1,4})'
        r'(?P<exp_year>[0-9]{2})'
        r'(?P<exp_day>[0-9]{2})'
        r'(?P<exp_month_opt_type>[A-Z])'
        r'(?P<strike>[0-9]+(?:\.[0-9]+)?)'
        r'(?P<extra>.*?)'  ## The remainder
        r'$'  ## Up to the end (explicit)
    ))

    """
    Example `(filename, data)`:
    ("GMCR1228L36.5_D.csv",
     {'exp_day': '28',
      'exp_month_opt_type': 'L',
      'exp_year': '12',
      'extra': '_D.csv',
      'strike': '36.5',
      'symbol': 'GMCR'})
    """

    ## http://www.iqfeed.net/symbolguide/index.cfm?symbolguide=guide&displayaction=support&section=guide&web=iqfeed&guide=options&web=IQFeed&type=stock
    exp_month_call = dict(
        (v, i + 1)
        for i, v in enumerate(
            ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L']))
    exp_month_put = dict(
        (v, i + 1)
        for i, v in enumerate([
            'M', 'N', 'O', 'P', 'Q', 'R', 'S', 'T', 'U', 'V', 'W', 'X']))

    def parse_exp_date(self, data):
        emot = data['exp_month_opt_type']
        if emot in self.exp_month_call:
            opt_type = 'C'
            exp_month_num = self.exp_month_call[emot]
        elif emot in self.exp_month_put:
            opt_type = 'P'
            exp_month_num = self.exp_month_put[emot]
        else:
            return self._mk_error('type_unmatched')

        exp_year = data['exp_year']
        if len(exp_year) == 2:
            exp_year = '20%s' % (exp_year,)
        exp_year = int(exp_year)

        exp_day = int(data['exp_day'])

        exp_orig_date = datetime.date(exp_year, exp_month_num, exp_day)
        exp_date = simple_expiration_date(
            exp_orig_date, skip_days=0, from_last_month_day=False)
        return dict(
            exp_date=exp_date,
            ## Well, that one is determined here
            opt_type=opt_type,
            ## Debug stuff
            exp_year_parsed=exp_year,
            exp_orig_date=exp_orig_date,
        )

    _columns = 'date time O H L C V OI'.split()
    _target_columns = 'O H L C V OI'.split()

    @classmethod
    def load_func(
            cls,
            filepath,
            ## Possibly wrong
            columns=None,
            ## Desired ordering (for output)
            target_columns=None,
            fileinfo=None):

        kwa1 = dict(
            filepath_or_buffer=filepath,
            parse_dates={'timestamp': [0, 1]},
            index_col='timestamp',
            date_parser=lambda v: datetime.datetime.strptime(v, '%d.%m.%Y %H%M%S'),
            header=None,
            names=columns or cls._columns,
        )
        try:
            df = pa.read_csv(**kwa1)
        except TypeError:
            df = pa.read_csv(**dict(
                kwa1,
                parse_dates=0,
                names='timestamp O H L C V OI'.split(),
                date_parser=lambda v: datetime.datetime.strptime(v, '%d.%m.%Y')
            ))

        df = pa.DataFrame(df, columns=target_columns or cls._target_columns)
        # df = ohlcv_resample(df)
        return df

    ## Shouldn't be used:
    @classmethod
    def load_func_ung(
            cls,
            filepath,
            ## Possibly wrong
            ## Need 'X' because of the ',' at the end of each line
            ## (apparently, pandas counts the column from the right)
            columns='O H L C V OI X'.split(),
            ## Desired ordering (for output)
            target_columns=None,
            fileinfo=None):
        df = pa.read_csv(
            filepath,
            parse_dates=True,
            index_col=0,
            date_parser=lambda v: datetime.datetime.strptime(v, '%Y-%m-%d %H:%M:%S'),
            header=None,
            # names=columns or cls._columns,
        )
        df.index.name = 'timestamp'
        df.columns = columns
        # df = pa.DataFrame(df, columns=target_columns or cls._target_columns)
        return df


if _run_dbg:
#     print pretty(OptParserCBOE().parse("GMCR1228L36.5_D.csv"))
#     print ""
#     print OptParserCBOE.load_func(base_dir + 'CBOE/GMCR/1W/GMCR1311A40.5_1.csv').head(3)
    ## ...
    print pretty(OptParserCBOE().parse("UNG1530M16.5.csv"))
    print ""
    # dfx = OptParserCBOE.load_func(base_dir + 'CBOE/GMCR/1M/OI/GMCR1319J55_D.csv')
    # dfx = OptParserCBOE.load_func(base_dir + 'CBOE/UNG/UNG1530M16.5.csv')
    dfx = OptParserCBOE.load_func(base_dir + 'CBOE/SPY/SPY_Weekly/SPY1227G137_D.csv')
    print dfx.head(3)
```

```python


def nearest_point(lst, val, presort=True):
    ## This might help to force rounding of 1.5 to 2 and such
    if presort:
        lst_prepared = sorted(lst, reverse=True)
    res = min(lst_prepared, key=lambda v: abs(v - val))
    return res


def nearest_higher(lst, val):
    return nearest_point((v for v in lst if v >= val), val)


def nearest_lower(lst, val):
    return nearest_point((v for v in lst if v <= val), val)



class NoData(Exception):
    """ Exception to be raised when there's no relevant data to be obtained """


def get_options(
        symbol, date, price, opt_type, distance,
        midstrike_rounding=True,
        debug=0,
):
    """ ...
    
    :param opt_type: 'P' or 'C'.
    :param midstrike_rounding: consider the nearest strike the midstrike rather than only the matching one.
    """
    strike_midprice = ftype(price)
    db = load_metadata_cached()
    ## symbol -> exp_date -> opt_type -> strike

    symbol_db = db[symbol]
    
    next_expiry = None
    try:
        mgr = symbol_db['__manager']
    except KeyError:
        pass
    else:
        try:
            next_expiry = mgr.nearest_expiration_date(date)
        except NotImplementedError:
            pass
    ## XX: '>='? Unlikely.
    ## TODO: optimise all this stuff before optimising the parameters
    if next_expiry is None:
        relevant_dates = [
            d for d in symbol_db
            if isinstance(d, (datetime.date, datetime.datetime))
            and d > date]
        if not relevant_dates:
            raise NoData("No relevant dates", None, symbol, date)
        next_expiry = min(relevant_dates)
    if debug >= 3:
        print "I: selected option with expiry %r" % (next_expiry,)
    date_db = symbol_db.get(next_expiry, {})
    if not date_db:
        raise NoData("Empty date_db", None, symbol, date, next_expiry)
    dir_db = date_db.get(opt_type, {})
    if not dir_db:
        raise NoData("Empty dir_db", None, symbol, date, next_expiry, opt_type)
    ## XXX: '=='?
    # higher_strikes, lower_strikes = split_list(dir_db.keys(), lambda v: v >= strike_midprice)
    matching_strikes = [v for v in dir_db if v == strike_midprice]
    if not matching_strikes:
        if midstrike_rounding:  ## Configurable
            strike_midprice_prev = strike_midprice
            round_func = nearest_point
            
            if midstrike_rounding == 'up':
                round_func = nearest_higher
            elif midstrike_rounding == 'down':
                round_func = nearest_lower
            elif midstrike_rounding == 'by_type':
                round_func = (
                    nearest_higher if opt_type == 'C'
                    else nearest_lower)

            if debug >= 3:
                print "D: round_func: %r" % (round_func,)
            strike_midprice = round_func(dir_db, strike_midprice)
            matching_strikes = [strike_midprice]
            if debug >= 1:
                print "I: rounding the strike from %r to %r" % (
                    strike_midprice_prev, strike_midprice)
        elif debug:
            print "W: no matching strikes", strike_midprice
    higher_strikes = [v for v in dir_db if v > strike_midprice]
    lower_strikes = [v for v in dir_db if v < strike_midprice]
    _sort_strikes = lambda l: sorted(l, key=lambda v: abs(v - strike_midprice))
    ## (!!!) this is for using the available options only for the distance
    ## rather than counting by the strike step from the midprice
    higher_strikes = _sort_strikes(higher_strikes)
    lower_strikes = _sort_strikes(lower_strikes)
    if debug >= 2:
        _lmt = debug * 3
        print "D: lower_strikes:", lower_strikes[:_lmt]
        print "D: matching_strikes:", matching_strikes[:_lmt]
        print "D: higher_strikes:", higher_strikes[:_lmt]

    if distance > 0:
        strikes = higher_strikes
    elif distance < 0:
        strikes = lower_strikes
    else:
        strikes = matching_strikes
    dist_idx = abs(distance - 1) if distance != 0 else 0
    try:
        strike = strikes[dist_idx]
    except IndexError:
        raise NoData("No relevant strike", sys.exc_info(), strikes, distance, dist_idx)

    fileinfo = dir_db[strike]  ## Shouldn't really fail
    if debug >= 2:
        print "loading file:", fileinfo.get('file_path')
        if debug >= 3:
            print "fileinfo:", fileinfo
    data = load_from_fileinfo_cached(fileinfo)

    if debug >= 2:
        print "data dates:", data.index

    ## (!!!) NOTE: assuming the actual decision moment is buy_date 23:59:59,
    ## i.e. get option data from the next date
    buy_date = date + datetime.timedelta(1)
    buy_date_check = simple_expiration_date(buy_date, 0, from_last_month_day=False, delta=1, debug=debug)
    try:
        data_start = data.ix[buy_date:].iloc[0]
    except IndexError:
        raise NoData("No option data from the the buy date", sys.exc_info(), buy_date, data.index, data)
    data_start_date = data_start.name.date()
    if data_start_date != buy_date and data_start_date != buy_date_check:
        if debug >= 2:
            ## NOTE: this might be normal if date + 1 is a weekend / holiday
            print("W: first option date != requested",
                  date, buy_date, buy_date_check, "...", data_start_date)
    exp_date = fileinfo['exp_date']
    ## NOTE: exp_date is supposed to be included
    try:
        data_exp = data.ix[:exp_date].iloc[-1]  ## WARN: raises IndexError if there's no data
    except IndexError:
        raise NoData("No data for the option expiration date", sys.exc_info(), exp_date, data.index, data)
    data_exp_date = data_exp.name.date()
    data_actual_end = data.iloc[-1].name
    if data_actual_end.date() != data_exp.name.date():
        if debug:
            print "W: last option date != option last pre-expiry date", data_actual_end, data_exp.name
    if data_exp_date != exp_date:
        if debug > 4:
            print "W: no data for option expiry date, last", data_exp_date, "required", exp_date

    return fileinfo, data_start, data_exp


if _run_dbg:
    ## This doesn't really work (no data error):
    # get_options(symbol='QNG', date=datetime.date(2009, 5, 30), price=4.5, opt_type='P', distance=1, debug=3)
    ## ...
    print get_options(symbol='QNG', date=datetime.date(2012, 12, 1), price=8, opt_type='C', distance=1, debug=3)
```

```python
def option_profit(symbol, date, *ar, **kwa):
    """ ...
    
    :param direction: `1` (buy) or `-1` (sell)
    :param return_all: return a dict with arious auxiliary info.
    
    Requires all parameters for `get_options`.
    """
    base_active = kwa.pop('base_active')  ## Currently required; remove later.
    return_all = kwa.pop('return_all', False)
    direction = kwa.pop('direction', 1)
    opt_type = kwa.get('opt_type')
    assert direction in (-1, 1)
    debug = kwa.get('debug', 0)
    optinfo, opt_a, opt_b = get_options(symbol=symbol, date=date, **kwa)
    close_date = optinfo['exp_date']
    ## XXX: use opt_b.name as close_date?
    ba_on_close = base_active.ix[:close_date].iloc[-1]  ## WARN: raises IndexError if there's no data
    if debug >= 4:
        print "ba_on_close:", ba_on_close
    ba_actual_date = ba_on_close.name.date()
    if debug >= 4:
        print "ba_actual_date:", ba_actual_date
    if close_date != ba_actual_date:
        if debug:
            print "W: found BA data date != close date", ba_actual_date, close_date

    opt_strike = optinfo['strike']
    ## NOTE: might want to turn all the other data into `ftype`s while loading
    opt_price = opt_a.O
    ba_price = ba_on_close.C
    if debug >= 2:
        print "opt_strike: ", opt_strike
        print "opt_price: ", opt_price
        print "ba_price: ", ba_price
        print "opt profit: ", ba_price - min(ba_price, opt_strike)
    if opt_type == 'C':
        profit = ba_price - min(ba_price, opt_strike) - opt_price
    elif opt_type == 'P':
        profit = max(ba_price, opt_strike) - ba_price - opt_price

    ## Re-check: sell_profit = -buy_profit
    # buy_profit = ba_price - min(ba_price, buy_strike) - buy_price
    # sell_profit = - (ba_price - min(ba_price, sell_strike) - sell_price)
    profit *= direction
    
    if return_all:
        return dict(
            profit=profit,
            close_date=close_date,
            optinfo=optinfo,
            opt_a=opt_a,
            opt_b=opt_b,
            ba_price=ba_price,
            ba_on_close=ba_on_close,
            opt_price=opt_price,
        )
    return profit


if _run_dbg:
    qng_ba = (
        pa.DataFrame.from_csv(base_dir + 'QNG#_D.csv')
        .sort_index()
        .rename(columns=dict(Open='O', High='H', Low='L', Settle='C', Volume='V')))
    ba_entry = qng_ba.ix['2012-12-01':].iloc[0]
    _r = option_profit(
        symbol='QNG', date=ba_entry.name.date(), price=ba_entry.C, opt_type='C',
        distance=1, debug=3, base_active=qng_ba)
    print ""
    print _r
```
